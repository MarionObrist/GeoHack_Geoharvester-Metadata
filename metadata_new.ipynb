```python
result = o.join(k)
```

```python
result
#print(type(result))
```

```python
grouped = result.groupby(['OWNER'])
grouped
num_kw = {}
for g in grouped:

    cells_filled = g[1].count()['KEYWORDS']
    num_cells = g[1].count()['OWNER']

    keyword_count = g[1]['KEYWORDS'].value_counts()
    num_corrupt_kw = keyword_count.get('[]')

    if num_corrupt_kw is not None:
        cells_filled -= num_corrupt_kw
    
    num_kw[g[0]] = [cells_filled, num_cells]

print(num_kw)
```

```python
frame = pd.DataFrame.from_dict(num_kw, orient='index',columns=['KEYWORDS','OWNER'])

frame['prozent'] = frame['KEYWORDS'] / frame['OWNER'] *100



frame
```

```python
frame.plot.bar()
```
